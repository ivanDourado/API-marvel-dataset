```python
import requests
import pandas as pd
```

```python
# API endpoint URL
url = "https://gateway.marvel.com/v1/public/characters"
# My Marvel API credentials
public_key = "62d5de58f13cfb840a014d9d8557543a"
private_key = "a54a74efaa8509a1c740fbeddd780799984bc77b"
```

```python
# import these libs for 
# Create an MD5 hash from a timestamp to include in request parameters
import hashlib
import time

ts = str(time.time())
```

```python
# Define the parameters for the API request
params = {
    "apikey": public_key,
    "ts": ts,
    "hash": hashlib.md5((ts + private_key + public_key).encode("utf-8")).hexdigest(),
    "limit": 100, # Set the number of results per page
    "offset": 0 # Set the offset to start retrieving results from
}
```

```python
# Send a GET request to the API endpoint and retrieve the response
response = requests.get(url, params=params)
#data = response.json()["data"]["results"]
total_characters = response.json()["data"]["total"]
```

```python
print(total_characters)
```

```python
# Create an empty list to store the data
results = []
```

```python
# Iterate through the pages of results and append to the results list
for i in range(0, total_characters, 100):
    # Update the offset parameter for the next page of results
    params["offset"] = i
# Send a GET request to the API endpoint with the updated parameters
    response = requests.get(url, params=params)
# Extract the data from the response JSON and append to the results list
    data = response.json()["data"]["results"]
    results.extend(data)
```

### To get all the data, I had to make several requests to the API, using the offset parameter to get the characters in blocks of fixed size, until I got all the desired results.

```python
# Convert the results to a Pandas DataFrame
df = pd.DataFrame(results, columns=["id", "name", "description", "comics", "series", "stories", "events"])
```

```python
# Write the DataFrame to a CSV file
df.to_csv("marvel_characters.csv", index=False)
```
